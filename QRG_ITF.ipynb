```python
import numpy as np
import matplotlib.pyplot as plt

delta = np.linspace(0, 2, 100)

def D(Del):
    d_n = Del**2
    return d_n

D_n1 = [D(q_val) for q_val in delta]
D_n2 = [D(q_val1) for q_val1 in D_n1]
D_n3 = [D(q_val2) for q_val2 in D_n2]
D_n4 = [D(q_val3) for q_val3 in D_n3]
D_n5 = [D(q_val4) for q_val4 in D_n4]
D_n6 = [D(q_val5) for q_val5 in D_n5]
D_n7 = [D(q_val6) for q_val6 in D_n6]
D_n8 = [D(q_val7) for q_val7 in D_n7]
D_n9 = [D(q_val8) for q_val8 in D_n8]
D_n10 = [D(q_val9) for q_val9 in D_n9]

def Cl(Del):
    q = Del + np.sqrt(1 + Del**2)
    c = (2*abs(q))/(1 + q**2)
    return c

def Cr(Del):
    q = Del + np.sqrt(1 + Del**2)
    cr = (-2*q**2/(1 + q**2)) * (np.log2(q)) + np.log2(1 + q**2)
    return cr

cl_values = [Cl(q_val) for q_val in delta]
cl_values1 = [Cl(d) for d in D_n1]
cl_values2 = [Cl(d2) for d2 in D_n2]
cl_values3 = [Cl(d3) for d3 in D_n3]
cl_values4 = [Cl(d4) for d4 in D_n4]
cl_values5 = [Cl(d5) for d5 in D_n5]
cl_values6 = [Cl(d6) for d6 in D_n6]
cl_values7 = [Cl(d7) for d7 in D_n7]
cl_values8 = [Cl(d8) for d8 in D_n8]
cl_values9 = [Cl(d9) for d9 in D_n9]
cl_values10 = [Cl(d10) for d10 in D_n10]


cr_values = [Cr(q_val) for q_val in delta]
cr_values1 = [Cr(d) for d in D_n1]
cr_values2 = [Cr(d2) for d2 in D_n2]
cr_values3 = [Cr(d3) for d3 in D_n3]
cr_values4 = [Cr(d4) for d4 in D_n4]
cr_values5 = [Cr(d5) for d5 in D_n5]
cr_values6 = [Cr(d6) for d6 in D_n6]
cr_values7 = [Cr(d7) for d7 in D_n7]
cr_values8 = [Cr(d8) for d8 in D_n8]
cr_values9 = [Cr(d9) for d9 in D_n9]
cr_values10 = [Cr(d10) for d10 in D_n10]

# Plot c vs delta
#plt.plot(delta, cl_values, label='0th iteration')
plt.plot(delta, cl_values1, label='1st iteration')
#plt.plot(delta, cl_values2, label='2nd iteration')
plt.plot(delta, cl_values3, label='3rd iteration')
#plt.plot(delta, cl_values4, label='4th iteration')
plt.plot(delta, cl_values5, label='5th iteration')
#plt.plot(delta, cl_values6, label='6th iteration')
plt.plot(delta, cl_values7, label='7th iteration')
#plt.plot(delta, cl_values8, label='8th iteration')
plt.plot(delta, cl_values9, label='9th iteration')
#plt.plot(delta, cl_values10, label='10th iteration')
plt.xlabel('$λ$')
plt.ylabel('$C_{l_1}$')
plt.title()
plt.grid(False)
plt.legend()
plt.show()

#plt.plot(delta, cr_values, label='0th iteration')
plt.plot(delta, cr_values1, label='1st iteration')
#plt.plot(delta, cr_values2, label='2nd iteration')
plt.plot(delta, cr_values3, label='3rd iteration')
#plt.plot(delta, cr_values4, label='4th iteration')
plt.plot(delta, cr_values5, label='5th iteration')
#plt.plot(delta, cr_values6, label='6th iteration')
plt.plot(delta, cr_values7, label='7th iteration')
#plt.plot(delta, cr_values8, label='8th iteration')
plt.plot(delta, cr_values9, label='9th iteration')
#plt.plot(delta, cr_values10, label='10th iteration')
plt.xlabel('$λ$')
plt.ylabel('$C_{RE}$')
plt.title()
plt.grid(False)
plt.legend()
plt.show()


cl_derivative = np.gradient(cl_values, delta)
cl_derivative1 = np.gradient(cl_values1, delta)
cl_derivative2 = np.gradient(cl_values2, delta)
cl_derivative3= np.gradient(cl_values3, delta)
cl_derivative4= np.gradient(cl_values4, delta)
cl_derivative5= np.gradient(cl_values5, delta)
cl_derivative6= np.gradient(cl_values6, delta)
cl_derivative7= np.gradient(cl_values7, delta)
cl_derivative8= np.gradient(cl_values8, delta)
cl_derivative9= np.gradient(cl_values9, delta)
cl_derivative10= np.gradient(cl_values10, delta)

cr_derivative = np.gradient(cr_values, delta)
cr_derivative1 = np.gradient(cr_values1, delta)
cr_derivative2 = np.gradient(cr_values2, delta)
cr_derivative3= np.gradient(cr_values3, delta)
cr_derivative4= np.gradient(cr_values4, delta)
cr_derivative5= np.gradient(cr_values5, delta)
cr_derivative6= np.gradient(cr_values6, delta)
cr_derivative7= np.gradient(cr_values7, delta)
cr_derivative8= np.gradient(cr_values8, delta)
cr_derivative9= np.gradient(cr_values9, delta)
cr_derivative10= np.gradient(cr_values10, delta)

#plt.plot(delta, cl_derivative, label='0th iteration')
plt.plot(delta, cl_derivative1, label='1st iteration')
#plt.plot(delta, cl_derivative2, label='2nd iteration')
plt.plot(delta, cl_derivative3, label='3rd iteration')
#plt.plot(delta, cl_derivative4, label='4th iteration')
plt.plot(delta, cl_derivative5, label='5th iteration')
#plt.plot(delta, cl_derivative6, label='6th iteration')
plt.plot(delta, cl_derivative7, label='7th iteration')
#plt.plot(delta, cl_derivative8, label='8th iteration')
plt.plot(delta, cl_derivative9, label='9th iteration')
#plt.plot(delta, cl_derivative10, label='10th iteration')
plt.xlabel('$λ$')
plt.ylabel('$dC_{l_1}/dλ$')
plt.title()
plt.grid(False)
plt.legend()
plt.show()


#plt.plot(delta, cr_derivative, label='0th iteration')
plt.plot(delta, cr_derivative1, label='1st iteration')
#plt.plot(delta, cr_derivative2, label='2nd iteration')
plt.plot(delta, cr_derivative3, label='3rd iteration')
#plt.plot(delta, cr_derivative4, label='4th iteration')
plt.plot(delta, cr_derivative5, label='5th iteration')
#plt.plot(delta, cr_derivative6, label='6th iteration')
plt.plot(delta, cr_derivative7, label='7th iteration')
#plt.plot(delta, cr_derivative8, label='8th iteration')
plt.plot(delta, cr_derivative9, label='9th iteration')
#plt.plot(delta, cr_derivative10, label='10th iteration')
plt.xlabel('$λ$')
plt.ylabel('$dC_{RE}/dλ$')
plt.title()
plt.grid(False)
plt.legend()
plt.show()
```
