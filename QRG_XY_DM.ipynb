```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def D(Del, M):
    return (3*Del + Del*3*M**2 + Del**3)/(1 + M**2 + 3*Del**2)

def Cl(Del, M):
    q = np.sqrt(1 + Del**2 + M**2)
    return (np.sqrt(2)*abs(Del)/q) + (np.sqrt(2)*(1+M**2)/q) +(abs(Del)*(1+M**2)/(q**2))

def Cr(Del, M):
    q = np.sqrt(1 + Del**2 + M**2)
    l = (Del**2)/(2*q**2)
    t = (1+M**2)/(2*q**2)
    return 1 - (l*np.log2(l)) - (t*np.log2(t))

# Values of M to iterate over
M_values = np.linspace(0, 2, 10)  # Define M values

# Lambda values
delta = np.linspace(-0.4, 0.4, 100)

# Iterate over M values
for M in M_values:
    D_n1 = [D(q_val, M) for q_val in delta]
    D_n2 = [D(q_val1, M) for q_val1 in D_n1]
    D_n3 = [D(q_val2, M) for q_val2 in D_n2]
    D_n4 = [D(q_val3, M) for q_val3 in D_n3]
    D_n5 = [D(q_val4, M) for q_val4 in D_n4]
    D_n6 = [D(q_val5, M) for q_val5 in D_n5]
    D_n7 = [D(q_val6, M) for q_val6 in D_n6]
    D_n8 = [D(q_val7, M) for q_val7 in D_n7]
    D_n9 = [D(q_val8, M) for q_val8 in D_n8]
    D_n10 = [D(q_val9, M) for q_val9 in D_n9]

# Create a meshgrid of M and delta
M_grid, delta_grid = np.meshgrid(M_values, delta)

# Compute Cl for each combination of M and delta
cl_values = np.array([[Cl(q_val, M_val) for q_val in D_n5] for M_val in M_values])

# Compute Cr for each combination of M and delta
cr_values = np.array([[Cr(q_val, M_val) for q_val in D_n5] for M_val in M_values])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(delta_grid, M_grid, cl_values.T, cmap='inferno')

# Set labels and title
ax.set_xlabel('$\gamma$')
ax.set_ylabel('$D$')
ax.set_zlabel('$C_{l_1}$')
ax.set_title('$C_{l_1}$ for different $\gamma$ and $D$ values')

plt.show()

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Plot the surface
ax.plot_surface(delta_grid, M_grid, cr_values.T, cmap='inferno')

# Set labels and title
ax.set_xlabel('$\gamma$')
ax.set_ylabel('$D$')
ax.set_zlabel('$C_{RE}$')
ax.set_title('$C_{RE}$ for different $\gamma$ and $D$ values')

plt.show()
```
